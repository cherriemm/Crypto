## 实验内容

### 使用蒙哥马利模幂算法的 RSA

参考[Courrège et. al, 2010]中选择明文攻击的条件:

- 选择有明确汉明重的message，例如 $M$ = 0000010000...00000 i.e. $M = 2^i$ ， $i$ 小于 modulus N. 

- 采集侧信道曲线（一条或者几条）

- 可以根据单条侧信道曲线，分析出RSA模幂运算过程中在做平方还是乘法。模幂运算的乘法比平方有更低的操作相关值（即更小的纵坐标）


### 参考文献

**[Courrège et al., 2010]** Courrège, J., Feix, B., and Roussellet, M. (2010). Simple
power analysis on exponentiation revisited. In Gollmann, D., Lanet, J., and
Iguchi-Cartigny, J., editors, _Smart Card Research and Advanced Application,
9th IFIP WG 8.8/11.2 International Conference, CARDIS 2010, Passau,
Germany, April 14-16, 2010._ Proceedings, volume 6035 of _Lecture Notes in
Computer Science_, pages 65–79. Springer.

参考给定的划分方法，对轨迹图形进行划分，区分模幂运算的不同操作

```python
plt.plot(trace)

plt.plot([21800 for _ in range(-80,30)],[i for i in range(-80,30)],color = 'r')
plt.plot([81600 for _ in range(-80,30)],[i for i in range(-80,30)],color = 'r')
plt.plot([99100 for _ in range(-80,30)],[i for i in range(-80,30)],color = 'y')
plt.plot([139000 for _ in range(-80,30)],[i for i in range(-80,30)],color = 'y')

plt.show()
```

已知：

1.模幂运算的乘法比平方有更低的操作相关值（即更小的纵坐标）

2.模幂运算，对于指数的 bit =0 时，只有平方运算； bit=1 时，平方运算之后会有乘法运算

根据曲线图像，我们可以分析出d的值。验证攻击结果是否正确。

```python
m = 128
d = 97673
n = 7753
s = pow(128,d,7753)
s
```

```python
def binary_to_decimal(binary_num):
    decimal_num = 0
    power = 0
    for digit in (binary_num):
        decimal_num += int(digit) * (2 ** power)
        power += 1
    return decimal_num

binary = "10010001101111101" #reversed
decimal = binary_to_decimal(binary)
print(decimal)
```
